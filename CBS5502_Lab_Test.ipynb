### **📝 Question 1**
Generate word embeddings with provided corpus for '*Universe*' and '*Earth*' using Word2Vec in random 10-dimensional vectors and compute the result of '*Universe*' - '*Earth*'.

Hints:
- create your own corpus by reading '*corpus1.txt*';
- train your model from gensim.models import Word2Vec and feed the corpus as training data;
- set key parameter n=10, other parameters are optional and train the model;
- get word embeddings for both words;
- compute 'Universe' - 'Earth' and display the result.


```python
# Q1: Word2Vec embeddings for 'Universe' and 'Earth' (10 dimensions)
from gensim.models import Word2Vec
from pathlib import Path
import re
import numpy as np

# Read corpus1.txt (fallback corpus is used only if file not found)
corpus_path = Path('corpus1.txt')
if corpus_path.exists():
    text = corpus_path.read_text(encoding='utf-8', errors='ignore')
else:
    text = """
    The Universe is vast and full of stars.
    The Earth is a planet in the Universe.
    We study Earth and the Universe in science.
    """

# Build corpus: tokenize into words (lowercase)
sentences = []
for line in text.splitlines():
    tokens = re.findall(r"[A-Za-z]+", line.lower())
    if tokens:
        sentences.append(tokens)

# Train Word2Vec with 10-dimensional random-initialized vectors
model = Word2Vec(
    sentences=sentences,
    vector_size=10,
    window=5,
    min_count=1,
    workers=1,
    sg=1,
    seed=42
)

universe_vec = model.wv['universe']
earth_vec = model.wv['earth']
result = universe_vec - earth_vec

print("Universe embedding:", universe_vec)
print("Earth embedding:", earth_vec)
print("Universe - Earth:", result)
print("Vector shape:", result.shape)
```

### **📝 Question 2**
Utilize Forward Maximum Matching (FMM) to do word segmentation for the sentence "學生會期望在課室内開派對" with the given dictionary:

word_list = {
    "學生",
    "會",
    "朋友",
    "舉辦",
    "期望",
    "在",
    "課室",
    "内",
    "開",
    "派對",
    "開派對",
    "一起",
    "幸福快樂",
    "活動",
    "計劃",
}

```python
# Q2: Forward Maximum Matching (FMM)
word_list = {
    "學生", "會", "朋友", "舉辦", "期望", "在", "課室", "内", "開", "派對",
    "開派對", "一起", "幸福快樂", "活動", "計劃"
}
sentence = "學生會期望在課室内開派對"

def fmm_segment(text, dictionary):
    max_len = max(len(w) for w in dictionary)
    i = 0
    result = []

    while i < len(text):
        matched = None
        # Try longest word first
        for L in range(min(max_len, len(text) - i), 0, -1):
            cand = text[i:i+L]
            if cand in dictionary:
                matched = cand
                break
        if matched is None:
            matched = text[i]  # fallback: single char
        result.append(matched)
        i += len(matched)

    return result

segmented = fmm_segment(sentence, word_list)
print("Original sentence:", sentence)
print("FMM result:", segmented)
```

### **📝 Question 3**
Predict the next token of the given sequence “明天不上” based on the given training corpus "*corpus2.txt*" using the **bi-gram** model.


```python
# Q3: Predict next token for "明天不上" using a bi-gram model
from pathlib import Path
from collections import Counter, defaultdict

corpus_path = Path('corpus2.txt')
if corpus_path.exists():
    text = corpus_path.read_text(encoding='utf-8', errors='ignore')
else:
    text = "明天不上課。明天不上班。今天不上學。明天不下雨。"

# Character-level tokens (remove spaces/newlines)
tokens = [ch for ch in text if not ch.isspace()]

# Build bigram counts: count(prev -> next)
bigram_counts = defaultdict(Counter)
for prev_ch, next_ch in zip(tokens[:-1], tokens[1:]):
    bigram_counts[prev_ch][next_ch] += 1

sequence = "明天不上"
last_token = sequence[-1]

if last_token in bigram_counts and len(bigram_counts[last_token]) > 0:
    next_token, freq = bigram_counts[last_token].most_common(1)[0]
    print("Input sequence:", sequence)
    print("Predicted next token:", next_token)
    print("Count:", freq)
    print("Top candidates:", bigram_counts[last_token].most_common(5))
else:
    print("No bigram found for token:", last_token)
```

### **📝 Question 4**
Given the dataset where x = [23, 25, 30, 35, 40, 45, 50, 55, 60, 65] (age) is the independent variable and y = [25, 30, 35, 40, 42, 45, 50, 53, 55, 60] (income in thousands) is the dependent variable, calculate:

- The Pearson correlation coefficient to measure the linear relationship between x and y.
- The fitted linear regression equation using the least squares method.


```python
# Q4: Pearson correlation and least-squares linear regression
import numpy as np

x = np.array([23, 25, 30, 35, 40, 45, 50, 55, 60, 65], dtype=float)
y = np.array([25, 30, 35, 40, 42, 45, 50, 53, 55, 60], dtype=float)

# 1) Pearson correlation coefficient
pearson_r = np.corrcoef(x, y)[0, 1]

# 2) Least-squares line: y = b0 + b1*x
b1, b0 = np.polyfit(x, y, 1)  # returns slope, intercept

print(f"Pearson correlation coefficient (r): {pearson_r:.4f}")
print(f"Fitted linear regression equation: y = {b0:.4f} + {b1:.4f}x")
```

### **📝 Question 5**
Given the following confusion matrix for a binary classification model:

    Confusion Matrix:
                     Predicted: 0    Predicted: 1
    Actual: 0           50              10
    Actual: 1           5               35

Write a Python program to calculate the following evaluation metrics:

- Accuracy
- Precision
- Recall
- F1-score


```python
# Q5: Evaluation metrics from confusion matrix
# TN=50, FP=10, FN=5, TP=35
TN, FP, FN, TP = 50, 10, 5, 35

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
```
