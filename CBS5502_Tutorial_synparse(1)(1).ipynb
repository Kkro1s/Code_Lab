## 📝 The Example Sentence

> **“He saw the man with a telescope.”**

This sentence is famous in linguistics and NLP because it has **two valid syntactic interpretations**:

1. 🔭 *He used a telescope to see the man* (PP attaches to **VP**)  
2. 👔 *The man has a telescope* (PP attaches to **NP**)

Our parser will help us **explicitly represent** these possibilities.

---


## 🧩 **1. Define a CFG Grammar**

In this step, we define a **Context-Free Grammar (CFG)** that captures the possible syntactic structures of our example sentence:

> **“He saw the man with a telescope.”**

This grammar is intentionally designed to **allow structural ambiguity**, so that our parser can generate **multiple valid parse trees**.


```python
"""
===========================================================
Top-Down Parsing with CFG using NLTK
===========================================================

This script demonstrates, STEP BY STEP, how to:

1. Define a Context-Free Grammar (CFG)
2. Avoid left recursion for top-down parsing
3. Initialize a Recursive Descent (top-down) parser
4. Parse a structurally ambiguous sentence
5. Observe multiple parse trees produced by the parser

Example sentence:
    "He saw the man with a telescope"

This is a CLASSIC linguistically ambiguous sentence:
- Did *He* use a telescope?
- Or does *the man* have a telescope?

-----------------------------------------------------------
"""

# ---------------------------------------------------------
# STEP 0: Import Required Libraries
# ---------------------------------------------------------
# nltk: Natural Language Toolkit (core NLP library)
# CFG: Used to define context-free grammars
# RecursiveDescentParser: A top-down parser implementation
# ---------------------------------------------------------

import nltk
from nltk import CFG
from nltk.parse import RecursiveDescentParser


# ---------------------------------------------------------
# STEP 1: Define the Context-Free Grammar (CFG)
# ---------------------------------------------------------
# IMPORTANT:
# - Top-down parsers CANNOT handle left-recursive rules
#   (e.g., NP -> NP PP).
# - Therefore, we rewrite the grammar to eliminate
#   left recursion while preserving ambiguity.
#
# This grammar intentionally allows TWO PP attachment sites:
#   1) NP -> DT NN PP  (PP attaches to the noun phrase)
#   2) VP -> V NP PP   (PP attaches to the verb phrase)
# ---------------------------------------------------------

grammar = CFG.fromstring("""
    S  -> NP VP

    # Noun Phrase rules
    NP -> PRP
    NP -> DT NN
    NP -> DT NN PP

    # Verb Phrase rules
    VP -> V NP
    VP -> V NP PP

    # Prepositional Phrase rule
    PP -> IN NP

    # Lexical rules
    PRP -> 'He'
    V   -> 'saw'
    DT  -> 'the' | 'a'
    NN  -> 'man' | 'telescope'
    IN  -> 'with'
""")

# Display the grammar
print("=== Context-Free Grammar ===")
print(grammar)
print("\n")


# ---------------------------------------------------------
# STEP 2: Define the Sentence to Parse
# ---------------------------------------------------------
# The parser expects a LIST of tokens (words), not a string.
# ---------------------------------------------------------

sentence = "He saw the man with a telescope".split()

print("=== Input Sentence ===")
print(sentence)
print("\n")


# ---------------------------------------------------------
# STEP 3: Initialize the Top-Down Parser
# ---------------------------------------------------------
# The RecursiveDescentParser:
# - Starts from the start symbol S
# - Expands grammar rules top-down
# - Matches terminals with input words
# - Backtracks when a rule fails
# ---------------------------------------------------------

parser = RecursiveDescentParser(grammar)


# ---------------------------------------------------------
# STEP 4: Parse the Sentence
# ---------------------------------------------------------
# parser.parse(sentence) returns ALL valid parse trees
# that can generate the sentence using the grammar.
# ---------------------------------------------------------

print("=== Parsing the Sentence (Top-Down Parsing) ===\n")

parse_count = 0

for tree in parser.parse(sentence):
    parse_count += 1

    # Print the bracketed tree structure
    print(f"Parse Tree {parse_count}:")
    print(tree)
    print()

    # Pretty-print the tree in a hierarchical layout
    print("Visualized Parse Tree:")
    tree.pretty_print()
    print("-" * 50)

# ---------------------------------------------------------
# STEP 5: Summary of Results
# ---------------------------------------------------------
# If the grammar is working correctly, you should see:
# - TWO valid parse trees
# - One where the PP attaches to the VP
# - One where the PP attaches to the NP
# ---------------------------------------------------------

print(f"\nTotal number of valid parses found: {parse_count}")

if parse_count > 1:
    print("✅ Structural ambiguity detected!")
else:
    print("⚠️ Only one parse found — ambiguity may be missing.")


# ---------------------------------------------------------
# KEY TAKEAWAYS (for students)
# ---------------------------------------------------------
# 1. CFGs can naturally represent syntactic ambiguity.
# 2. Top-down parsers are intuitive but inefficient.
# 3. Parsing does NOT choose the 'correct' meaning.
# 4. Disambiguation requires probabilities, semantics,
#    or world knowledge.
#
# Try experimenting by:
# - Removing VP -> V NP PP
# - Changing the sentence
# - Adding adjectives or new nouns
# ---------------------------------------------------------
```

## 🧠 **Step‑by‑Step Explanation of the Output**

---

### 🧩 **Sentence Structure**

The parser analyzes the sentence as a hierarchical structure:

- **Noun Phrase (NP)**  
  - *“He”*  
  - Classified as a **pronoun (PRP)**

- **Verb Phrase (VP)**  
  - **Verb (V):** *“saw”*  
  - **Noun Phrase (NP):** *“the man with a telescope”*

---

### 🔍 **Internal Structure of the Object NP**

The noun phrase *“the man with a telescope”* is further decomposed:

- **Core NP**
  - **DT:** *“the”*
  - **NN:** *“man”*

- **Prepositional Phrase (PP)**
  - **IN:** *“with”*
  - **NP:** *“a telescope”*
    - **DT:** *“a”*
    - **NN:** *“telescope”*

This structure allows the grammar to represent complex noun phrases.

---

### 🔄 **Top‑Down Parsing Process**

- The parser starts from the root symbol **S**
- Grammar rules are applied **recursively from top to bottom**
- Each expansion attempts to match the next word in the input sentence
- As rules succeed, the parse tree is built incrementally
- Backtracking is used when a rule fails, ensuring all valid parses are explored

---

## 📝 **Notes & Practical Considerations**

### ⚙️ **Grammar Customization**
- The grammar can be extended to handle richer linguistic phenomena:
  - adjectives (*the tall man*)
  - adverbs (*saw quickly*)
  - more complex phrase structures
- Simply add or modify production rules as needed

---

### 🌈 **Handling Structural Ambiguity**
- Top‑down parsing may produce **multiple valid parse trees**
- In the sentence:
  > *“He saw the man with a telescope”*
  
  the phrase *“with a telescope”* may modify:
  - ✅ the **verb** (*how he saw*)
  - ✅ the **noun** (*which man*)

- The parser systematically explores **all grammatically valid interpretations**

---

### ⚠️ **Error Handling**
- If the grammar does not contain the necessary rules:
  - ❌ the sentence cannot be parsed
  - ❌ no parse tree will be produced

---

### ✅ **Key Takeaway**
> **Parsing reveals structure, not meaning.**  
CFG‑based parsers expose ambiguity explicitly, highlighting the need for probabilities, semantics, or context to resolve interpretation.

### 📝 **Extended Solution: Resolving Ambiguity with PCFGs**

To go beyond basic CFG parsing, we use **Probabilistic Context‑Free Grammars (PCFGs)** to resolve **structural ambiguity** by ranking parses with probabilities.

---

## 🌟 Why PCFGs?
- CFGs generate **all possible parses**
- PCFGs assign **probabilities to grammar rules**
- The parser selects the **most likely parse tree**

✅ This reflects how NLP systems prefer *plausible* interpretations.

---

## 🧠 Parsing Setup
- **Grammar:** PCFG  
- **Parser:** `nltk.ChartParser`  
- **Goal:** Choose the highest‑probability parse

---

## 🧪 Example Sentence
> **“He saw the man with a telescope.”**

The PP *“with a telescope”* may attach to:
- 🔭 the **verb** (*how he saw*)
- 👔 the **noun** (*which man*)

PCFGs help prefer one interpretation over the other.

---

➡️ **Next:** Implement PCFG parsing in Python and inspect parse probabilities.

```python
"""
===========================================================
PCFG Parsing with NLTK — Resolving Structural Ambiguity
===========================================================

This example demonstrates how to use a
Probabilistic Context-Free Grammar (PCFG)
together with the Viterbi Parser to:

• assign probabilities to grammar rules
• parse an ambiguous sentence
• select the MOST LIKELY parse tree

Example sentence:
    "He saw the man with a telescope"

-----------------------------------------------------------
"""

import nltk
from nltk import PCFG
from nltk.parse import ViterbiParser


# ---------------------------------------------------------
# STEP 1: Define the PCFG Grammar
# ---------------------------------------------------------
# IMPORTANT RULE:
# For each non-terminal, the probabilities of its productions
# MUST sum to 1.0.
#
# These probabilities encode our linguistic preferences:
# - VP -> V NP is more likely than VP -> V NP PP
# - DT NN is the most common NP structure
# - NP PP is less frequent but still possible
# ---------------------------------------------------------

pcfg_grammar = PCFG.fromstring("""
    S  -> NP VP                [1.0]

    NP -> PRP                  [0.2]
    NP -> DT NN                [0.5]
    NP -> NP PP                [0.3]

    VP -> V NP                 [0.6]
    VP -> V NP PP              [0.4]

    PP -> IN NP                [1.0]

    PRP -> 'He'                [1.0]
    V   -> 'saw'               [1.0]
    DT  -> 'the'               [0.8] | 'a' [0.2]
    NN  -> 'man'               [0.6] | 'telescope' [0.4]
    IN  -> 'with'              [1.0]
""")

print("=== Probabilistic Grammar ===")
print(pcfg_grammar)
print()


# ---------------------------------------------------------
# STEP 2: Tokenize the Input Sentence
# ---------------------------------------------------------
# The parser expects a list of tokens (words)
# ---------------------------------------------------------

sentence = "He saw the man with a telescope".split()

print("=== Input Sentence ===")
print(sentence)
print()


# ---------------------------------------------------------
# STEP 3: Initialize the Viterbi Parser
# ---------------------------------------------------------
# The Viterbi parser:
# - is a bottom-up chart parser
# - efficiently explores all parses
# - returns the HIGHEST-PROBABILITY parse
# ---------------------------------------------------------

parser = ViterbiParser(pcfg_grammar)


# ---------------------------------------------------------
# STEP 4: Parse the Sentence
# ---------------------------------------------------------
# Unlike CFG parsing, PCFG parsing typically yields
# ONE best parse (the most probable one).
# ---------------------------------------------------------

print("=== Parsing with PCFG (Viterbi Algorithm) ===\n")

for tree in parser.parse(sentence):
    print("Most Likely Parse Tree (with probability):")
    print(tree)
    print()

    print("Visualized Parse Tree:")
    tree.pretty_print()
    print("-" * 50)


# ---------------------------------------------------------
# KEY OBSERVATIONS
# ---------------------------------------------------------
# • The PP "with a telescope" attaches to the VP or NP
#   depending on rule probabilities.
# • Changing probabilities changes the preferred parse.
# • PCFGs move us from *possible* parses to *plausible* parses.
#
# Try experimenting:
# - Increase NP -> NP PP probability
# - Decrease VP -> V NP PP probability
# ---------------------------------------------------------
```

This sentence has **two valid syntactic interpretations**:

## 💡 Hint

> **PCFGs don’t eliminate ambiguity — they rank it.**  
Changing probabilities changes which meaning the parser prefers.

Understand how **probabilistic grammar design** directly influences syntactic interpretation.

```python
# Exercise solution: "He fed her cat food."
from nltk import PCFG
from nltk.parse import ViterbiParser

pcfg_ex = PCFG.fromstring("""
S  -> NP VP                 [1.0]
NP -> PRP                   [0.35]
NP -> PRP NN                [0.30]
NP -> NN NN                 [0.25]
NP -> PRP NN NN             [0.10]
VP -> V NP                  [1.0]

PRP -> 'He'                 [0.5] | 'her' [0.5]
V   -> 'fed'                [1.0]
NN  -> 'cat'                [0.5] | 'food' [0.5]
""")

tokens = "He fed her cat food".split()
parser = ViterbiParser(pcfg_ex)
parses = list(parser.parse(tokens))

print("Number of parses:", len(parses))
for i, t in enumerate(parses, 1):
    print(f"\nParse {i} | prob={t.prob():.6f}")
    print(t)

print("\nInterpretation note:")
print("- Parse A: her cat (possessor reading)")
print("- Parse B: cat food (compound noun reading)")
print("- Viterbi selects the parse with higher rule-product probability.")
```
